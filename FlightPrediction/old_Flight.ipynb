```python
test_df = pd.read_excel('Test_set.xlsx')
test_df.head()
```

```python
final_df = train_df.append(test_df)
final_df.tail() ### Disadvantage of using append  is the index values will be repeated from test and train datasets
```

```python
final_df.info()
```

This dataset has date and time objects, so we convert these columns in to date time objects and derive different features like day, month, year etc

```python
final_df['Date_of_Journey'].str.split('/').str[0]
```

```python
###Feature Engineering process
##HEre we create three columns as date month and year and get the info from date_of_jounrny by converting the object into string and splitting
final_df['Date'] = final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month'] = final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year'] = final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df.info()
```

```python
final_df['Month']=final_df['Month'].astype(int)
final_df['Date']=final_df['Date'].astype(int)
final_df['Year']=final_df['Year'].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop('Date_of_Journey', axis = 1, inplace = True)
```

```python
final_df.head(1)
```

```python
## Arrival_time
# IF we see the arrival_time in some colums there is some redundant information like date and month 
## So we can remove that information and just get the time 
final_df['Arrival_Time']=final_df['Arrival_Time'].apply(lambda x:x[0:5])
```

```python
final_df.head(1)
```

```python
final_df['Arrival_hour'] = final_df['Arrival_Time'].str.split(':').str[0]
final_df['Arrival_min'] = final_df['Arrival_Time'].str.split(':').str[1]
```

```python
final_df.info()
```

```python
final_df['Arrival_hour']=final_df['Arrival_hour'].astype(int)
final_df['Arrival_min']=final_df['Arrival_min'].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop('Arrival_Time', axis = 1, inplace = True)
```

```python
final_df.head()
```

```python
##Now we do the same process for the departure time
final_df['Dep_hour'] = final_df['Dep_Time'].str.split(':').str[0]
final_df['Dep_min'] = final_df['Dep_Time'].str.split(':').str[1]
```

```python
final_df['Dep_hour']=final_df['Dep_hour'].astype('int')
final_df['Dep_min']=final_df['Dep_min'].astype('int')
```

```python
final_df.drop('Dep_Time', axis = 1, inplace = True)
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
##Route
## in this column we get infomation about the names of the stops and total number of stops between the departure and destination
##But we already have another column which gives the info about number of stops and there is no use of knowing what are those stops
## so we can simply drop the route column
## Before droping this column look for more information from route and total_stops columns
```

```python
final_df['Total_Stops'].unique()
```

```python
final_df['Total_Stops'].isnull().sum()
```

```python
final_df[final_df['Total_Stops'].isnull()]
## For delhi to cochin we need 2 stops(got by looking into other info)
```

```python
## By using map
final_df['Total_Stops'] =final_df['Total_Stops'].map({'non-stop' : 0,
                                                      '1 stop': 1,
                                                      '2 stops': 2,
                                                       '3 stops': 3,
                                                       '4 stops': 4,
                                                      'NaN': 2})
```

```python
final_df.head(5)
```

```python
final_df.drop('Route', axis = 1, inplace =True)
```

```python
final_df.info()
```

```python
##Duration
final_df['Duration_hrs']=final_df['Duration'].apply(lambda x:x.split(' ')[0]).apply(lambda x:x.replace('h',''))
```

```python
final_df['Duration_hrs'].unique()
## After cplitting the duration into hr and min, we are facing an issue to convert the hour column into int data type
## SO, by looking into the uinque values, we can see that there is 5m entry which can not be converted into a int
```

```python
final_df[final_df['Duration_hrs']=='5m']
## we should drop these 2 column as there is some misinfomation in these columns( we can not travel from mumbai to hyd in 5 min)
```
